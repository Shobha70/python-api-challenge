Obervations:
    1. Temperature is inversely related to latitude. As latitude increases, the temperature falls, and vice versa. It looks like        at this time of the year, temperature is more higher in southern hemisphere
    2. Humidity is generally higher near the equator.
    3. There seems to be a weak correlation between latitude and cloudiness
    4. Wind Speed generally increases slowly with latitude
    

```python
#import libraries
import random
import numpy as np
import pandas as pd

from citipy import citipy
from pprint import pprint
import requests
from config import api_key

import matplotlib.pyplot as plt
from scipy.stats import linregress

from datetime import date 

import logging
```

```python
# select random latitude ranges
latitude_range = np.random.uniform(-90,90,size=1500)
```

```python
# select random longitude ranges
longitude_range = np.random.uniform(-180,180,size=1500)
```

```python
# create a new dataframe with a random list of latitude and longitude ranges
coordinate_df = pd.DataFrame({'latitude':latitude_range,
                               'longitude':longitude_range})
```

```python
#build url for open weather map api and set units to "imperial"
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# Build partial query URL
query_url = f"{url}appid={api_key}&units={units}&q="
```

```python
#build cities dataframe by using citipy library and randomly selected latitude and longitudes
cities = []
lat = []
lng = []
for i in range(0, len(coordinate_df['latitude'])):
    city = citipy.nearest_city(coordinate_df.iloc[i]['latitude'],coordinate_df.iloc[i]['longitude'])
    city_name = city.city_name
    response = requests.get(query_url + city_name).json()
#     pprint(response)
    if response['cod'] == 200:
        if city_name not in cities:
            cities.append(city_name)
            lat.append(coordinate_df.iloc[i]['latitude'])
            lng.append(coordinate_df.iloc[i]['longitude'])
        
cities_df = pd.DataFrame({'city': cities,
                           'latitude': lat,
                           'longitude': lng}) 
```

```python
# append temperature, humidity, cloudiness, and wind speed to cities dataframe.
# find temperature, humidity, cloudiness, and windspeed from open weather api
temp = []
hum = []
cloud = []
ws = []

# Include a print log of each city as it's being processed with the city number and city name.
#Create and configure logger 
logging.basicConfig(filename="cities.log", 
                    format='%(asctime)s %(message)s', 
                    filemode='w') 
  
#Creating an object 
logger=logging.getLogger() 
  
#Setting the threshold of logger to INFO 
logger.setLevel(logging.INFO) 



for i in range(0, len(cities_df['city'])):
    response = requests.get(query_url + cities_df.iloc[i]['city']).json()
#     pprint(response)
           
    temp.append(response['main']['temp'])
    hum.append(response['main']['humidity'])
    cloud.append(response['clouds']['all'])
    ws.append(response['wind']['speed'])
    
#   log of each city as it's being processed
    logger.info("City:" + cities_df.iloc[i]['city'] + ",latitude:" + str(cities_df.iloc[i]['latitude']) + ",Temperature(F):" + str(response['main']['temp']) + ",Humidity(%):" + str(response['main']['humidity']) + ",Cloudiness(%):" + str(response['clouds']['all']) + ",Wind Speed(mph):" + str(response['wind']['speed'])) 
    
    
    
cities_df['temperature(F)'] = temp
cities_df['humidity(%)'] = hum
cities_df['cloudiness(%)'] = cloud
cities_df['wind speed(mph)'] = ws

logging.shutdown()
# cities_df
```

```python
# Save a CSV of all retrieved data
cities_df.to_csv ('cities.csv', index = False, header=True)
```

```python
# create a series of scatter plots to showcase the following relationships:

# * Temperature (F) vs. Latitude
# * Humidity (%) vs. Latitude
# * Cloudiness (%) vs. Latitude
# * Wind Speed (mph) vs. Latitude

x_col = "latitude"
y_columns = ["temperature(F)", "humidity(%)", "cloudiness(%)","wind speed(mph)"]


for y_col in y_columns:

    figure = plt.figure
    ax = plt.gca()
    ax.scatter(cities_df[x_col], cities_df[y_col],edgecolors='black')
    ax.set_xlabel(x_col)
    ax.set_ylabel(y_col)
    ax.set_title("{} vs {} - {}".format(x_col, y_col, date.today().strftime("%m/%d/%Y")))
    ax.grid()

    if y_col == 'temperature(F)':
        plt.text(-100, -100, "Temperature decreases as you move away from the equator")
    elif y_col == 'humidity(%)':
        plt.text(-100, -40, "Humidity appears to be mostly regular across latitudes\n with an absence lower humidity near equator and artic")
    elif y_col == 'cloudiness(%)':
        plt.text(-100, -30, "Cloudiness does not seem to vary significantly over latitudes")
    elif y_col == 'wind speed(mph)':    
        plt.text(-100, -15, "Wind Speed shows a slow increase as you move away from equator")
      
    plt.savefig('scatterplot '+ x_col +' vs '+ y_col +'.png')
    plt.show() 
```

```python
# Create a dataframe for Northern Hemisphere (greater than or equal to 0 degrees latitude)
nh_df = cities_df[cities_df['latitude']>=0]
sh_df = cities_df[cities_df['latitude']<0]
```

```python
# run linear regression on each relationship for Northern hemisphere
x_col = "latitude"

y_columns = ["temperature(F)", "humidity(%)", "cloudiness(%)","wind speed(mph)"]


for y_col in y_columns:

#     figure = plt.figure
#     ax = plt.gca()
#     ax.scatter(nh_df[x_col], nh_df[y_col])
#     ax.set_xlabel(x_col)
#     ax.set_ylabel(y_col)
    
    fig, (ax1, ax2) = plt.subplots(2)
    
    ax1.scatter(nh_df[x_col], nh_df[y_col],edgecolors='black')
#     ax1.set_title('Northern Hemisphere')
    ax1.set_xlabel(x_col)
    ax1.set_ylabel(y_col)
    ax2.scatter(sh_df[x_col], sh_df[y_col],edgecolors='black' )
#     ax2.set_title('Southern Hemisphere')
    ax2.set_xlabel(x_col)
    ax2.set_ylabel(y_col)
    
   
    
    #y = mx + b for Northern Hemisphere
    x_val = nh_df[x_col].tolist()
    y_val = nh_df[y_col].tolist()
    
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x_val,y_val)
    
    y = []
    for index in range(len(nh_df[x_col])):
        y.append(nh_df.iloc[index][x_col].tolist() * slope + intercept)
       
    ax1.plot(x_val,y,'r-')
    
    ax1.set_title(f'Northern-The correlation between {x_col} and {y_col} is ' + str(round(rvalue,2)))
    
#   y = mx + b for Southern Hemisphere
    x_val = sh_df[x_col].tolist()
    y_val = sh_df[y_col].tolist()
    
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x_val,y_val)
    
    y = []
    for index in range(len(sh_df[x_col])):
        y.append(sh_df.iloc[index][x_col].tolist() * slope + intercept)
       
    ax2.plot(x_val,y,'r-')
    
    ax2.set_title(f'Southern-The correlation between {x_col} and {y_col} is ' + str(round(rvalue,2)))
    
    fig.tight_layout()
    plt.subplots_adjust(top=.85)
    fig.suptitle(f'{x_col} vs {y_col}')
    
    if y_col == 'temperature(F)':
        plt.text(-100, -30, "Temperature and latitude have a strong inverse correlation.")
    elif y_col == 'humidity(%)':
        plt.text(-100, -40, "Humidity appears to be mostly regular across latitudes\n with an absence lower humidity near equator and artic")
    elif y_col == 'cloudiness(%)':
        plt.text(-100, -60, "Cloudiness does seem to have a weak correlation to latitude")
    elif y_col == 'wind speed(mph)':    
        plt.text(-100, -15, "Wind Speed shows a slow increase as you move away from equator")


    plt.savefig('Linear Regression -  '+ x_col +' vs '+ y_col +'.png')
    plt.show()
```
